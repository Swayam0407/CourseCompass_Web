```python
import pandas as pd
import numpy as np

# Load data from CSV file
path='sample_courses_dataset2.csv'
courses_df = pd.read_csv(path)

# Inspect the first few rows and unique values
print("First few rows of 'Certification (Yes/No)' column:")
print(courses_df[['Certification (Yes/No)']].head(10))

print("Unique values in 'Certification (Yes/No)':")
print(courses_df['Certification (Yes/No)'].unique())

# Check the data type
print("Data type of 'Certification (Yes/No)':", courses_df['Certification (Yes/No)'].dtype)

# Print raw representations of values to check for hidden characters
print("Raw representations of 'Certification (Yes/No)' values:")
print(courses_df['Certification (Yes/No)'].apply(lambda x: repr(x)).unique())

# Convert 'Certification (Yes/No)' to string, strip spaces, and upper case
courses_df['Certification (Yes/No)'] = courses_df['Certification (Yes/No)'].astype(str).str.strip().str.upper()

# Print unique values after conversion
print("Unique values in 'Certification (Yes/No)' after cleaning:")
print(courses_df['Certification (Yes/No)'].unique())

# Handle conversion using replace() method
certification_map = {'YES': 1, 'NO': 0}
courses_df['Certification (Yes/No)'] = courses_df['Certification (Yes/No)'].replace(certification_map)

# Fill NaN values with 0
courses_df['Certification (Yes/No)'] = courses_df['Certification (Yes/No)'].fillna(0)

# Verify the changes
print("Updated 'Certification (Yes/No)' column:")
print(courses_df[['Certification (Yes/No)']].head(10))

# Print value counts to verify the distribution
print("Value counts of 'Certification (Yes/No)':")
print(courses_df['Certification (Yes/No)'].value_counts(dropna=False))

# Optional: Check data type after conversion
print("Data type of 'Certification (Yes/No)' after mapping:", courses_df['Certification (Yes/No)'].dtype)

courses_df.head()
```

```python
import numpy as np

# weights
weights = {
    "Rating": 0.25,
    "Number of Reviews": 0.15,
    "Enrollment Count": 0.15,
    "Completion Rate": 0.10,
    "Instructor Reputation": 0.15,
    "Time Spent on Course Materials (hours)": 0.05,
    "Update Frequency (months)": 0.05,
    "Certification (Yes/No)": 0.05,
    "Price ($)": 0.05
}

# Normalize
courses_df["Rating"] /= 5
courses_df["Number of Reviews"] /= courses_df["Number of Reviews"].max()
courses_df["Enrollment Count"] /= courses_df["Enrollment Count"].max()
courses_df["Completion Rate"] /= 100
courses_df["Instructor Reputation"] /= 5
courses_df["Time Spent on Course Materials (hours)"] /= courses_df["Time Spent on Course Materials (hours)"].max()
courses_df["Update Frequency (months)"] = 1 - (courses_df["Update Frequency (months)"] / 12)
courses_df["Price ($)"] = 1 - (courses_df["Price ($)"] / courses_df["Price ($)"].max())

# weighted scores
courses_df["Composite Score"] = (
    courses_df["Rating"] * weights["Rating"] +
    courses_df["Number of Reviews"] * weights["Number of Reviews"] +
    courses_df["Enrollment Count"] * weights["Enrollment Count"] +
    courses_df["Completion Rate"] * weights["Completion Rate"] +
    courses_df["Instructor Reputation"] * weights["Instructor Reputation"] +
    courses_df["Time Spent on Course Materials (hours)"] * weights["Time Spent on Course Materials (hours)"] +
    courses_df["Update Frequency (months)"] * weights["Update Frequency (months)"] +
    courses_df["Certification (Yes/No)"] * weights["Certification (Yes/No)"] +
    courses_df["Price ($)"] * weights["Price ($)"]
)

# topic wise grouping
courses_df['Rank'] = courses_df.groupby('Category/Topic')['Composite Score'].rank(ascending=False)

# Acc. to topic and rank
courses_df = courses_df.sort_values(by=['Category/Topic', 'Rank'])

print(courses_df[['Course Title', 'Category/Topic', 'Composite Score', 'Rank']])
```

```python
def get_courses_by_topic(topic):
    required_df = courses_df[courses_df['Category/Topic'].str.contains(topic, case=False, na=False)]

    if required_df.empty:
        print(f"No courses found for the topic: {topic}")
    else:
        print(f"Courses related to the topic: {topic}")
        print(required_df[['Course Title', 'Category/Topic', 'Composite Score', 'Rank']])

# user input
user_topic = input("Enter the topic to search for related courses: ").strip()
get_courses_by_topic(user_topic)
```
